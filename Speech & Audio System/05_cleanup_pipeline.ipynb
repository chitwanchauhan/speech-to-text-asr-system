```python
from transformers import WhisperProcessor, WhisperForConditionalGeneration
import torch
import librosa
import pandas as pd
from tqdm import tqdm

BASE_DIR = "/content/drive/MyDrive/JoshTalks_ASR"

seg_train_path = BASE_DIR + "/03_processed/train_val_split/seg_train.csv"
df = pd.read_csv(seg_train_path)

# Take small subset (IMPORTANT)
df = df.head(50).copy()

device = "cuda" if torch.cuda.is_available() else "cpu"

processor = WhisperProcessor.from_pretrained("openai/whisper-small")
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-small")
model.to(device)
```

```python
df = df.head(50).copy()  # keep small

raw_preds = []

from tqdm import tqdm
import librosa

for _, row in tqdm(df.iterrows(), total=len(df)):
    audio, _ = librosa.load(row["audio_path"], sr=16000)

    inputs = processor(audio, sampling_rate=16000, return_tensors="pt").input_features.to(device)

    ids = model.generate(inputs)
    pred = processor.batch_decode(ids, skip_special_tokens=True)[0]

    raw_preds.append(pred)

df["raw_asr"] = raw_preds
df["reference"] = df["text"]

df.head()
```

```python
hindi_num_map = {
    "शून्य":0, "एक":1, "दो":2, "तीन":3, "चार":4,
    "पांच":5, "छह":6, "सात":7, "आठ":8, "नौ":9,
    "दस":10, "ग्यारह":11, "बारह":12, "तेरह":13,
    "चौदह":14, "पंद्रह":15, "बीस":20, "तीस":30,
    "चालीस":40, "पचास":50, "साठ":60, "सत्तर":70,
    "अस्सी":80, "नब्बे":90, "सौ":100, "हजार":1000
}

def normalize_numbers(text):
    words = text.split()
    result = []
    temp_num = 0

    for w in words:
        if w in hindi_num_map:
            val = hindi_num_map[w]

            if val == 100:
                temp_num = max(1, temp_num) * 100
            elif val == 1000:
                temp_num = max(1, temp_num) * 1000
                result.append(str(temp_num))
                temp_num = 0
            else:
                temp_num += val
        else:
            if temp_num != 0:
                result.append(str(temp_num))
                temp_num = 0
            result.append(w)

    if temp_num != 0:
        result.append(str(temp_num))

    return " ".join(result)
```

```python
def normalize_numbers_safe(text):
    if "-" in text:
        return text  # skip idioms like दो-चार

    return normalize_numbers(text)
```

```python
df["num_normalized"] = df["raw_asr"].apply(normalize_numbers_safe)
```

```python
test_sentences = [
    "मेरे पास दो किताबें हैं",
    "उसने तीन सौ पचास रुपए दिए",
    "एक हजार लोग आए थे",
    "पच्चीस छात्र पास हुए",
    "दस दिन बाद परीक्षा है",

    # edge cases
    "दो-चार बातें करनी हैं",
    "एक दिन ऐसा आएगा"
]

for s in test_sentences:
    print("\nRAW:", s)
    print("NORMALIZED:", normalize_numbers_safe(s))
```

```python
english_like_words = [
    "इंटरव्यू", "जॉब", "कंप्यूटर", "प्रॉब्लम",
    "सॉल्व", "फील", "ट्रिप", "फिल्म", "बटन",
    "फोन", "ड्राइवर", "मैनेजर", "ट्रेन", "बस"
]

def tag_english_words(text):
    words = text.split()
    tagged = []

    for w in words:
        if w in english_like_words:
            tagged.append(f"[EN]{w}[/EN]")
        else:
            tagged.append(w)

    return " ".join(tagged)
```

```python
df["english_tagged"] = df["num_normalized"].apply(tag_english_words)
```
